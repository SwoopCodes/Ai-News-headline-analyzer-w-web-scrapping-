```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

news_headlines = pd.read_csv('./Training_Data/news_training_data.csv')
X = news_headlines["headline"]
Y = news_headlines["outcome"]

model = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", DecisionTreeClassifier())
])

model.fit(X, Y)

test_headlines = [
        "Tech Giant Faces Total Collapse As Product Launch Is Delayed",
    "Coca-Cola Sales Plunge Amid Worsening Middle East Supply Chain Crisis",
    "Nvidia Stock Crashes 12% After Disappointing Q1 Earnings Miss",
    "Amazon to Lay Off 10,000 Employees as Profit Margins Shrink",
    "Starbucks Closing 300 US Stores Following Steep Decline in Foot Traffic",
    "Alphabet Stock Hits New All-Time High on Strong AI Revenue Growth",
    "Coca-Cola Beats Q2 Estimates, Raises Full-Year Guidance",
    "Netflix Adds 15 Million Subscribers, Stock Surges 8%",
    "JPMorgan Declares Special Dividend After Record Quarterly Profits",
    "Costco Expands Nationwide, Same-Store Sales Jump 12%"
]
predictions = model.predict(test_headlines)

predictionsArray = []
for headline, pred in zip(test_headlines, predictions):
    matchedHeadline = f"{pred.upper()}: {headline}"
    predictionsArray.append(matchedHeadline)

predictionsArray

    
```
